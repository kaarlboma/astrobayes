```python
import pandas as pd
import matplotlib.pyplot as plt

# Load beta samples
df = pd.read_csv("beta_samples.csv", header=None)
df.columns = [f'beta_{i}' for i in range(df.shape[1])]

# Plot histograms
df.plot(kind='hist', bins=50, alpha=0.7, subplots=True, layout=(df.shape[1], 1), figsize=(8, 2*df.shape[1]), legend=False)
plt.tight_layout()
plt.show()
```
